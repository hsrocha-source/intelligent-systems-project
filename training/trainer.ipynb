```python
csv = os.environ['DATASET_PATH']
dframe = pd.read_csv(csv)
```

Vou separar os vetores da tabela nas informações linguísticas que considero mais úteis para categorização: query, title e concatenated_tags, e somá-las (com espaços). Também removi os NaN do concatenated tags, substituindo-os por strings vazias

```python
query = np.array(dframe['query'])
title = np.array(dframe['title'])
concatenated_tags = np.array(dframe['concatenated_tags'])
category = np.array(dframe['category'])
```

O processo a seguir converte o vetor de strings correspondente às categorias de treinamento em um vetor de números naturais


```python
occupied_array = []
for cat in category:
    if cat not in occupied_array:
        occupied_array.append(cat)

int_array = []
for cat in category:
    int_ = occupied_array.index(cat)
    int_array.append(int_)
```

```python
int_array = np.array(int_array)
```

Como podemos observar pela ordem da lista a seguir, associamos os números [0,1,2,3,4,5] com ['Decoração', 'Papel e Cia', 'Outros', 'Bebê', 'Lembrancinhas', 'Bijuterias e Jóias']

```python

print(occupied_array)
```

```python
#Remover NaN das tags concatenadas
concatenated_tags = np.where(type(concatenated_tags) != str, '', concatenated_tags)
```

```python
texto = query + ' ' + title + ' ' + concatenated_tags
```

```python
print(texto)
```

Pipeline de treinamento com Sklearn e validação cruzada com 8 splits. Para extrair os atributos das listas de strings, utilizei o modelo mais simples, de saco de palavras (aqui denotado como CountVectorizer, pela biblioteca sklearn). O modelo escolhido foi o Perceptron,  particularmente pela boa razão entre simplicidade e eficiência. A biblioteca Sklearn não possui modelo de stopwords para português. Acredita-se que as stopwords não irão influenciar demasiadamente no desempenho do modelo, já que são infrequentes no texto das tabelas. Logo, elas não foram removidas.

O código a seguir foi adaptado do exemplo no github do professor (MAC5832-PROJECT). Foram trocados o modelo, a função KFold, as métricas e vários parâmetros do processamento de texto. Adaptar pedaços de códigos disponíveis livremente é prática comum no mundo da ciência de dados.

O sistema todo demora menos de 10 minutos para ser treinado na minha CPU (i7 4a geração)

As métricas utilizadas são F1, precisão, recall e acurácia

```python
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score
from sklearn.linear_model import Perceptron
from sklearn.pipeline import Pipeline

kf = KFold(n_splits=8)

acc_in = [0] * 10
acc_val = [0] * 10

f1_in = [0] * 10
f1_val = [0] * 10

prec_in = [0] * 10
prec_val = [0] * 10

rec_in = [0] * 10
rec_val = [0] * 10

# Abrir metrics.txt no modo r+ para excluir as
# métricas do experimento anterior de treinamento do modelo
metrics = open(os.environ['METRICS_PATH'], 'r+')
metrics.truncate(0)
metrics.close()
for idx, [train_index, test_index] in enumerate(kf.split(texto)):
    # Abrir metrics.txt no modo append para inserir as
    # métricas de cada iteração do treinamento do modelo
    metrics = open(os.environ['METRICS_PATH'], 'a')
    print('Validation set #%d' % (idx+1))
    metrics.write('Validation set #%d \n' % (idx+1))
    

    Xk_train, Xk_test = texto[train_index], texto[test_index]
    Yk_train, Yk_test = int_array[train_index], int_array[test_index]
    
    #Tokenizer

    model = Pipeline([('vectorizer', CountVectorizer(stop_words=None)), ('model', Perceptron())])

    model.fit(Xk_train, Yk_train)

    predK_train = model.predict(Xk_train)
    predK_test = model.predict(Xk_test)
    
    acc_in[idx] = accuracy_score(Yk_train, predK_train) 
    acc_val[idx] = accuracy_score(Yk_test, predK_test)
    
    f1_in[idx] = f1_score(Yk_train, predK_train, average='weighted')
    f1_val[idx] = f1_score(Yk_test, predK_test, average='weighted')
    
    prec_in[idx] = precision_score(Yk_train, predK_train, average='weighted')
    prec_val[idx] = precision_score(Yk_test, predK_test, average='weighted')
    
    rec_in[idx] = recall_score(Yk_train, predK_train, average='weighted')
    rec_val[idx] = recall_score(Yk_test, predK_test, average='weighted')
    
    print('Training set accuracy : ', acc_in[idx])
    print('Test set accuracy     : ', acc_val[idx])
    print('')
    
    print('Training set F1 : ', f1_in[idx])
    print('Test set F1     : ', f1_val[idx])
    print('')
    
    print('Training set precision : ', prec_in[idx])
    print('Test set precision     : ', acc_val[idx])
    print('')
    
    print('Training set recall : ', prec_in[idx])
    print('Test set recall     : ', prec_val[idx])
    print('')
    
    #Adiciona as métricas F1, acc, prec, recall
    metrics.write('Training set accuracy : {} \n'.format(acc_in[idx]))
    metrics.write('Training set F1 : {} \n'.format(f1_in[idx]))
    metrics.write('Training set prevalsion : {} \n'.format(prec_in[idx]))
    metrics.write('Training set recall : {} \n'.format(rec_in[idx]))
    
    metrics.write('Test set accuracy : {} \n'.format(acc_val[idx]))
    metrics.write('Test set F1 : {} \n'.format(f1_val[idx]))
    metrics.write('Test set precision : {} \n'.format(prec_val[idx]))
    metrics.write('Test set recall : {} \n'.format(rec_val[idx]))
    metrics.close()
    
```

```python
import pickle
# Salvar o modelo
with open(os.environ['MODEL_PATH'], 'wb') as fid:
    pickle.dump(model, fid)
```
